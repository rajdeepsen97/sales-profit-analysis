# Sales & Profit Analysis

```python
import pandas as pd
```

```python
# WE WILL DISPLAY THE CSV FILE WITH FIRST 5 ROWS
```

```python
df = pd.read_csv("Superstore.csv", encoding='latin1')
df.head()
```

```python
# WE WILL DISPLAY ALL THE COULUMN NAMES 
```

```python
print(df.columns)
```

## Data Cleaning

```python
# Step 1: Clean Columns (SOMETIMES COULUMNS NAMES HAVE EXTRA SPACES, WE WILL REMOVE THEM BY :-)
```

```python
df.columns = df.columns.str.strip()                   #------------> Avoid hidden space errors like 'Profit ' vs 'Profit'
```

## Profit Analysis

```python
# Step 2: Profit Analysis (VERY IMPORTANT)
```

```python
profit_by_category = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(profit_by_category)
```

## Profit Analysis

```python
# Step 3: Profit by Region
```

```python
profit_by_region = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print(profit_by_region)
```

## Loss Analysis

```python
# Step 4: Loss-making Areas (THIS IMPRESSES INTERVIEWERS)
```

```python
df[df['Profit'] < 0].groupby('Category')['Profit'].sum()
```

## Visualization

```python
# Step 5: Visualization (matplotlib)
```

```python
import matplotlib.pyplot as plt

profit_by_category.plot(kind='bar')
plt.title("Profit by Category")
plt.show()
```

## Profit by Category

```python
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Category")
plt.show()
```

## Sales by Region 

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Region")
plt.show()
```
