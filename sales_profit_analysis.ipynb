# Sales & Profit Analysis

## Objective
The goal of this project is to analyze sales and profit data to identify high-performing categories, regions, and areas of loss.

```python
import pandas as pd
```

```python
# WE WILL DISPLAY THE CSV FILE WITH FIRST 5 ROWS
```

```python
df = pd.read_csv("Superstore.csv", encoding='latin1')
df.head()
```

```python
# WE WILL DISPLAY ALL THE COULUMN NAMES 
```

```python
print(df.columns)
```

## Data Cleaning

Before performing analysis, the dataset needs to be cleaned to ensure accuracy. 
This includes handling missing values, correcting data types, and standardizing column names.
Clean data is essential for reliable insights and prevents errors during analysis.

```python
# Step 1: Clean Columns (SOMETIMES COULUMNS NAMES HAVE EXTRA SPACES, WE WILL REMOVE THEM BY :-)
```

```python
df.columns = df.columns.str.strip()                   #------------> Avoid hidden space errors like 'Profit ' vs 'Profit'
```

## Exploratory Data Analysis

The goal of this step is to understand overall sales distribution across different regions and categories. 
This helps identify high-performing areas and potential opportunities for growth.

## Profit Analysis Category

```python
profit_by_category = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(profit_by_category)
```

## Profit Analysis Region

```python
profit_by_region = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print(profit_by_region)
```

## Loss Analysis Category

```python
df[df['Profit'] < 0].groupby('Category')['Profit'].sum()
```

## Insight: Loss Analysis

Certain transactions result in negative profit, particularly in the Furniture category. 
This may be due to excessive discounts or high operational costs.

## Visualization

```python
import matplotlib.pyplot as plt
```

## Profit by Category

```python
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Category")
plt.show()
```

## Insight: Category Profitability

Technology is the most profitable category, suggesting higher margins and strong demand. 
On the other hand, Furniture shows significantly lower profit, indicating potential inefficiencies.

## Sales by Region 

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Region")
plt.show()
```

## Insight: Regional Performance

The analysis shows that the West region generates the highest sales and profit. 
This indicates strong market performance and customer demand in this region.

## Sales Distribution by Category

```python
df.groupby('Category')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales Distribution by Category")
plt.show()
```
